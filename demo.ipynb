## Set up PaCo

Clone the repository and install dependencies. It may take 1-2 minutes.

```python
%%bash
git clone https://github.com/PaCo-X/PaCo-demo.git
# install pointnet2_ops_lib
cd PaCo-demo/pointnet2_ops_lib && python setup.py install
# install required PyPI packages
cd ../build && pip install -r requirements.txt
# set permission for PolyFit
cd ../PolyFit && chmod +x polyfit
# copy utils.py to root directory
cd ../..  && cp PaCo-demo/utils.py utils.py
```
